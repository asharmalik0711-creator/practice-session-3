### Question 1:

Write a Python program that asks the user to guess a secret word.

Keep asking until the user enters the correct word ("python").

When guessed correctly, print "🎉 You guessed it!"

Expected Output:

Guess the word: java

Guess the word: ruby

Guess the word: python

🎉 You guessed it!

```python
secret_word = "python"
guess = input("guess the word: " )

while guess != secret_word :
    guess = input("guess the word: " )
    
    print("🎉 You guessed it!")
```

### Q2 Stop sending emails after 3 users (Using for loop and break)

users = ["Ali", "Sara", "John", "Zara", "Mike"]

Output:

Email sent to Ali

Email sent to Sara

Email sent to John

Limit reached. Stop sending emails.

```python
users = ["Ali", "Sara", "John", "Zara", "Mike"]

for index, user in enumerate(users):
    if index == 3:
        print("Limit reached. Stop sending emails.")
        break

    print("Email sent to", user)
```

### Q3. Stop monitoring temperature if critical temperature reached (Using for loop and break)

temps = [22, 25, 30, 45, 28] critical = 40

Ouput:

Temperature: 22

Temperature: 25

Temperature: 30

Critical temperature reached! Stop monitoring.

```python
temps = [22, 25, 30, 45, 28] 
critical = 40

for temp in temps:
    if temp >= critical:
        print("Critical temperature reached! Stop monitoring.")
        break

    print("Temprature: ",temp)
```

### Q4. Calculate total price of items in a shopping cart (Using For loop)

cart = [100, 250, 50, 300]

Output:

Total Price: 700

```python
cart = [100, 250, 50, 300]

total = 0

for price in cart:
    total = total + price


print("Total Price:",total)
```

### Q5. Skip customers with unpaid bills (Using For loop and continue)

customers = [ {"name": "Ali", "paid": True},

{"name": "Sara", "paid": False},

{"name": "John", "paid": True} ]

Output:

Access granted to: Ali

Access granted to: John

```python
customers = [ 
    {"name": "Ali", "paid": True},
    {"name": "Sara", "paid": False},
    {"name": "John", "paid": True} 
]

for customer in customers:
    if not customer["paid"]:
        continue

        
    print ("Access granted to:", customer["name"])

```

### Q7. Extract domain names from email list (Using for loop)

emails = ["ali@gmail.com", "sara@yahoo.com", "john@hotmail.com"]

Output:

domains: ['gmail.com', 'yahoo.com', 'hotmail.com']

```python
emails = ["ali@gmail.com", "sara@yahoo.com", "john@hotmail.com"]

domains = []

for email in emails:
    domain = email.split("@")[1]
    domains.append(domain)

print("domains:", domains)
```

### Q9:

Write a Python program that keeps asking the user to enter numbers and prints them.

The program should stop the loop if the user enters a negative number.

Use a while True loop with a break statement

Expected Output:

Enter a number (negative to stop): 5

You entered: 5

Enter a number (negative to stop): 10

You entered: 10

Enter a number (negative to stop): -3

Negative number entered. Loop stopped!

```python
while True:
    number = int(input("Enter a number (negative to stop): "))

    if number < 0:
        print("Negative number entered. Loop stopped!")
        break

    print("You entered:", number)
```

### Q10. Skip weekends while printing workdays (Using for loop and continue)

days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

Output:

Workday: Monday

Workday: Tuesday

Workday: Wednesday

Workday: Thursday

Workday: Friday

```python
days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
for day in days:
    if day == "Saturday" or day == "Sunday":
        continue

    print("Workday:", day)
    print("      ")
```
